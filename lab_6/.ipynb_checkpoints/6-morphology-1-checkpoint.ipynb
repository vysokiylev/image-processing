# 6. morphology-1

http://scikit-image.org/docs/dev/api/skimage.morphology.html

```python
import numpy as np
from skimage.io import imread, imshow
from skimage.morphology import (closing, opening, 
                                binary_closing, binary_opening,
                                binary_erosion, binary_dilation, 
                                erosion, dilation, 
                                diamond, disk, rectangle, square, star)

import matplotlib.pyplot as plt
from skimage import data
from sklearn.datasets import load_digits
from skimage import img_as_float
%matplotlib inline
```

```python
from plot import plot2, plot3, plot4
```

```python
digits = load_digits()
imgs = digits.images[0:10]
```

```python
img0 = imread('images/0.png', 1)
imshow(img0)
plt.show()
```

```python
plt.imshow(imgs[3])
plt.gray()
plt.show()
```

```python
plot3(star(3), star(5), star(10), 'radius 3', 'radius 5', 'radius 10')
```

```python
plot3(disk(3), disk(5), disk(10), 'radius 3', 'radius 5', 'radius 10')
```

```python
plot3(diamond(3), diamond(5), diamond(10), 'radius 3', 'radius 5', 'radius 10')
```

```python
plot3(rectangle(3, 3), rectangle(3, 5), rectangle(3,7), '3 3', '3 5', '3 7')
```

```python
img1 = imgs[0]
img2 = imgs[1]
img3 = imgs[2]
plt.imshow(img1)
plt.show()
```

## erosion/dilation

### disk(1)

```python
plot3(img0, binary_erosion(img0, disk(1)), binary_dilation(img0, disk(1)), \
      'source', 'binary erosion', 'binary dilation')
```

```python
plot3(img1, binary_erosion(img1, disk(1)), binary_dilation(img1, disk(1)), \
      'source', 'binary erosion', 'binary dilation')
```

```python
plot3(img1, erosion(img1, disk(1)), dilation(img1, disk(1)), \
      'source', 'erosion', 'dilation')
```

```python
plot3(img2, binary_erosion(img2, disk(1)), binary_dilation(img2, disk(1)), \
      'source', 'binary erosion', 'binary dilation')
```

```python
plot3(img2, erosion(img2, disk(1)), dilation(img2, disk(1)), \
      'source', 'erosion', 'dilation')
```

### square(2)

```python
plot3(img0, binary_erosion(img0, square(2)), binary_dilation(img0, square(2)), \
      'source', 'binary erosion', 'binary dilation')
```

### rectangle(1,2)

```python
plot3(img0, binary_erosion(img0, rectangle(1,2)), binary_dilation(img0, rectangle(1,2)), \
      'source', 'binary erosion', 'binary dilation')
```

## opening/closing

```python
plot3(img0, binary_opening(img0, disk(1)), binary_closing(img0, disk(1)), 'source', 'binary opening', 'binary closing')
```

```python
plot3(img1, binary_opening(img1, disk(1)), binary_closing(img1, disk(1)), \
      'source', 'binary opening', 'binary closing')
```

```python
plot3(img1, opening(img1, disk(1)), closing(img1, disk(1)), \
      'source', 'opening', 'closing')
```

```python
plot3(img2, binary_opening(img2), binary_closing(img2), \
      'source', 'binary opening', 'binary closing')
```

```python
plot3(img2, opening(img2), closing(img2), \
      'source', 'opening', 'closing')
```

```python
plot3(img3, opening(img3), closing(img3), 'source', 'opening', 'closing')
```
