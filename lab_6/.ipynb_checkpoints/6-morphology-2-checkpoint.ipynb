## Используя операция математической морфологии:
### 1. Найти внутренние и внешние границы изображений 1 – 3 (шум удалить).

```python
import numpy as np
from skimage.io import imread, imshow
from skimage.morphology import (closing, opening, 
                                binary_closing, binary_opening,
                                binary_erosion, binary_dilation, 
                                erosion, dilation, 
                                diamond, disk, rectangle, square, star)


from skimage.color import rgb2gray, rgb2yuv, yuv2rgb
import matplotlib.pyplot as plt
from skimage import data
from sklearn.datasets import load_digits
from skimage import img_as_float, img_as_ubyte
from skimage.filters import rank
%matplotlib inline
```

```python
from plot import plot2, plot3, plot4
```

```python
img1 = imread('images/1.png', 1)
img2 = imread('images/2.png', 1)
img3 = imread('images/3.png', 1)
```

```python
img1_erosion = erosion(img1, rectangle(6,6))
img1_erosion_dilation =  dilation(img1_erosion, rectangle(6,6))
img1_clean =  img1_erosion_dilation - img1_erosion
plot4(img1, img1_erosion, img1_erosion_dilation, img1_clean, \
      'source', 'erosion', 'erosion dilation', 'difference')
```

```python
img2_dilation = dilation(img2, rectangle(6,6))
img2_dilation_erosion = erosion(img2_dilation, rectangle(5,5))
img2_clean = img2_dilation - img2_dilation_erosion
plot4(img2, img2_dilation, img2_dilation_erosion, img2_clean, \
      'source', 'dilation', 'dilation erosion', 'difference')
```

```python
img3_erosion = erosion(img3, rectangle(4,4))
img3_erosion_dilation = dilation(img3_erosion, rectangle(5,5))
img3_clean = img3_erosion_dilation - img3_erosion
plot4(img3, img3_erosion, img3_erosion_dilation, img3_clean, \
      'source', 'erosion', 'erosion dilation', 'difference')
```

### 4. Определить возраст дерева по спилу.

```python
img11 = imread('images/11.jpg', 1)
img11_cut = img11[1610:1650, 285:1490]
img11_cut = img_as_ubyte(img11_cut)
```

```python
plt.figure(figsize=(5,5))
imshow(img11)
plt.show()
```

```python
plt.figure(figsize=(20,20))
imshow(img11_cut)
plt.show()
```

_В правой части изображения темные кольца имеют тот же цвет, что и светлые слева_

```python
# Equalization
img11_cut = rank.equalize(img11_cut, selem=disk(30))

plt.figure(figsize=(20,20))
imshow(img11_cut)
plt.show()
```

```python
img11_cut = img_as_float(img11_cut>173)

plt.figure(figsize=(20,20))
plt.gray()
plt.imshow(img11_cut)
plt.show()
```

```python
img11_cut = dilation(img11_cut, square(3))
plt.figure(figsize=(20,20))
plt.imshow(img11_cut)
plt.show()
```

```python
rows, cols = img11_cut.shape
```

```python
line = np.mean(img11_cut, axis = 0)
```

```python
img11_cut = np.array([line]*40)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(img11_cut)
plt.show()
```

```python
img11_cut = img_as_float(binary_erosion(img11_cut, square(2)))
plt.figure(figsize=(20,20))
plt.imshow(img11_cut)
plt.show()
```

```python
def count_series(X):
    threshold = 0.5
    line = []
    for i in X:
        if i > threshold:
            line.append(1)
        else:
            line.append(0)
    qcnt = 0
    tau = 0
    cnt1 = 0
    v = 0
    for i in range(cols - 1):
        if line[i] == line[i+1]:
            cnt1 += 1
        else:
            cnt1 = 0
            v +=1
        if cnt1 > tau:
            tau = cnt1

    print("Возраст дерева (количество серий): ", (v-3)//2)
    print("Максимальная ширина серии ", tau)
```

```python
count_series(line)
```
